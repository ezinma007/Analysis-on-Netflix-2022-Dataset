## 2. What year has the highest amount of Tv shows and Movies added to Netflix? Show on a bar Graph.

## to_datetime

```python
data['Date_N'] = pd.to_datetime(data['date_added']) #Creates a new column named "Date_N" and convert a column named "date_added" in the DataFrame from a string format to a datetime format.
```

```python
data.dtypes # Checking the data type of the newly created column [Date_N]
```

## dt.year.value_counts()

```python
year_counts= data['Date_N'].dt.year.value_counts()    #It counts the occurence of all individual years in date column.
year_counts
```

## Bar Graph

```python
plt.figure(figsize=(10,6)) # increase the figure size
sns.barplot(x=year_counts.index, y=year_counts.values, color='red') #show data on a bargraph
```
